```python
import h5py
import matplotlib.pyplot as plt
```

```python
f = h5py.File("test.hdf5")
f1 = h5py.File("test_1.hdf5")
```

```python
t = f["time"][()]
```

```python
for n in f["neurons"].keys():
    x = f["neurons"][n]["voltage"]["data"][()][0]
    x1 = f1["neurons"][n]["voltage"]["data"][()][0]
    plt.plot(t, x)
    plt.plot(t, x1)
    plt.show()
```
