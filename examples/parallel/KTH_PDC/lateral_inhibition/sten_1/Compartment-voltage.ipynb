```python
%matplotlib inline

import os
network_path = os.path.join("..", "networks", "sten_1")
network_file = os.path.join(network_path, "network-synapses-minimal.hdf5")
simulation_file_with_synapses = os.path.join(network_path, "simulation", "output-with-synapses-sten_1-A.hdf5")
simulation_file_no_synapses = os.path.join(network_path, "simulation", "output-no-synapses-sten_1-A.hdf5")
duration = 18
```

```python
from snudda.plotting.plot_neuron_voltage import PlotNeuronVoltage

snudda_data = "../../../../../../BasalGangliaData/data/"

import numpy as np

pnv = PlotNeuronVoltage(network_path=network_path,
                        network_file=network_file,
                        simulation_file=simulation_file_with_synapses,
                        snudda_data=snudda_data)
```

```python
time_bins = [[2, 3], [5, 6], [8, 9], [11, 12], [14, 15], [17, 18]]

# neuron_id_list = pnv.network_loader.get_neuron_id(include_virtual=False)[:10]
neuron_id_list = [0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000, 7500]

sim_center = np.array([0.00475, 0.004, 0.00775])

for neuron_id in neuron_id_list:
    fig_name = os.path.join(network_path, "figures", f"dendritic_voltage-{neuron_id}-with-synapses-sten-1A.png")
    
    p_id = pnv.network_loader.get_neuron_population_units(neuron_id=neuron_id)[0]
    pos = (pnv.network_loader.data["neurons"][neuron_id]["position"] - sim_center) * 1e6
    
    pnv.plot_binned_neuron_voltage(neuron_id=neuron_id, time_bins=time_bins, title=f"With synapses (Pop {p_id}, pos {np.around(pos,1)})", 
                                   fig_name=fig_name, y_range=(-0.1, -0.01),
                                   fig_size=(10,10))
```

```python
from snudda.plotting.plot_neuron_voltage import PlotNeuronVoltage

snudda_data = "../../../../../../BasalGangliaData/data/"

pnv2 = PlotNeuronVoltage(network_path=network_path,
                         network_file=network_file,
                         simulation_file=simulation_file_no_synapses,
                         snudda_data=snudda_data)
```

```python
time_bins = [[2, 3], [5, 6], [8, 9], [11, 12], [14, 15], [17, 18]]

# neuron_id_list2 = pnv2.network_loader.get_neuron_id(include_virtual=False)[:10]
neuron_id_list2 = [0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000, 7500]

for neuron_id in neuron_id_list2:

    fig_name = os.path.join(network_path, "figures", f"dendritic_voltage-{neuron_id}-no-synapses-sten-1A.png")

    p_id = pnv2.network_loader.get_neuron_population_units(neuron_id=neuron_id)[0]
    pos = (pnv2.network_loader.data["neurons"][neuron_id]["position"] - sim_center) * 1e6
    
    pnv2.plot_binned_neuron_voltage(neuron_id=neuron_id, time_bins=time_bins, title=f"No synapses (Pop {p_id}, pos {np.around(pos,1)})", 
                                    fig_name=fig_name, y_range=(-0.1, -0.01),
                                    fig_size=(10,10))
```
