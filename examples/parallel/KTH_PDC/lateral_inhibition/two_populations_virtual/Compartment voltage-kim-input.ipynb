```python
%matplotlib inline

import os
network_path = os.path.join("..", "networks", "lateral_twopop_virtual")
network_file = os.path.join(network_path, "network-synapses-virtual.hdf5")
simulation_file_with_synapses = os.path.join(network_path, "simulation", "output-with-synapses-12Hz-virtual-kim.hdf5")
simulation_file_no_synapses = os.path.join(network_path, "simulation", "output-no-synapses-12Hz-virtual-kim.hdf5")
duration = 5
```

```python
from snudda.plotting.plot_neuron_voltage import PlotNeuronVoltage

snudda_data = "../../../../../../BasalGangliaData/data/"

pnv = PlotNeuronVoltage(network_path=network_path,
                        network_file=network_file,
                        simulation_file=simulation_file_with_synapses,
                        snudda_data=snudda_data)
```

```python
time_bins = [[0, 0.5], [0.5, 1], [1, 1.5], [1.5, 2], [2, 2.5], [2.5, 3], [3, 3.5], [3.5, 4], [4, 4.5], [4.5, 5], [5, 5.5]]

# neuron_id_list = pnv.network_loader.get_neuron_id(include_virtual=False)[:10]
neuron_id_list = [15563, 15784, 15801, 15687, 15881, 15704, 15777, 15732, 15845, 15643]

for neuron_id in neuron_id_list[:5]:
    fig_name = os.path.join(network_path, "figures", f"dendritic_voltage-{neuron_id}-with-synapses-5Hz-virtual.png")
    
    pnv.plot_binned_neuron_voltage(neuron_id=neuron_id, time_bins=time_bins, title="With synapses, 5Hz virtual neighbours", 
                                   fig_name=fig_name, y_range=(-0.1, -0.05),
                                   fig_size=(10,10))
```

```python
from snudda.plotting.plot_neuron_voltage import PlotNeuronVoltage

snudda_data = "../../../../../../BasalGangliaData/data/"

pnv2 = PlotNeuronVoltage(network_path=network_path,
                         network_file=network_file,
                         simulation_file=simulation_file_no_synapses,
                         snudda_data=snudda_data)
```

```python
time_bins = [[0, 0.5], [0.5, 1], [1, 1.5], [1.5, 2], [2, 2.5], [2.5, 3], [3, 3.5], [3.5, 4], [4, 4.5], [4.5, 5], [5, 5.5]]

# neuron_id_list2 = pnv2.network_loader.get_neuron_id(include_virtual=False)
neuron_id_list2 = [15563, 15784, 15801, 15687, 15881, 15704, 15777, 15732, 15845, 15643]


for neuron_id in neuron_id_list2[:5]:

    fig_name = os.path.join(network_path, "figures", f"dendritic_voltage-{neuron_id}-no-synapses-5Hz-virtual.png")

    pnv2.plot_binned_neuron_voltage(neuron_id=neuron_id, time_bins=time_bins, title=f"No synapses, 5Hz virtual neighbours", 
                                    fig_name=fig_name, y_range=(-0.1, -0.05),
                                    fig_size=(10,10))
```
