# DSPN neuromodulation

This jupyter notebook analyses ```Dardel_simulate_dspn_DA_bath.job```.

````python
import os
network_path = os.path.join("..", "networks", "dspn_DA_bath")
network_simulation_path = os.path.join(network_path, "simulation", "dspn-output.hdf5")

from snudda.utils import SnuddaLoadSimulation, SnuddaLoad

sl = SnuddaLoad(network_path)
sls = SnuddaLoadSimulation(network_path=network_path, network_simulation_output_file=network_simulation_path)
````

````python
from snudda.plotting import SnuddaPlotSpikeRaster2
spsr = SnuddaPlotSpikeRaster2(snudda_load=sl, snudda_simulation_load=sls, network_path=None)
spsr.plot_spike_raster()
````

````python
from snudda.plotting import PlotTraces
pt = PlotTraces(output_file=network_simulation_path, network_file=network_path)
````

````python
pt.plot_traces(fig_size=(12, 15), mark_spikes=False)
````

````python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(time, data_da)
plt.xlabel("Time (s)")
plt.ylabel("Concentration")
plt.title("Dopamine")
# plt.legend()
fig_path = os.path.join(network_path, "figures", "DA_figure.png")
plt.savefig(fig_path)
plt.show()
````

````python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(time, data_pka, label="PKAc")
plt.xlabel("Time (s)")
plt.ylabel("Concentration")
#plt.legend()
plt.title("PKAc")
fig_path = os.path.join(network_path, "figures", "PKAc_figure.png")
plt.savefig(fig_path)
plt.show()
````

````python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(time, data_kir_mod)
plt.xlabel("Time (s)")
plt.ylabel("Concentration")
plt.title("Kir modulation factor")
# plt.legend()
fig_path = os.path.join(network_path, "figures", "Kir_mod-figure.png")
plt.savefig(fig_path)
plt.show()
````
