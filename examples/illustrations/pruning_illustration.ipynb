## f1 and mu2 pruning

Combination of ```f1``` and ```mu2``` pruning. Here $P = f_1 / (1 + \exp(-8 / \mu_2 \cdot (n_{synapses} - \mu_2)))$, where $P$ is the probability to keep a synapse, $n_{synapses}$ is the total number of synapses connecting the neuron pre-post pair.

### Config files:
* [network-config-f1-1-mu2-3.json](pruning_illustration_network/network-config-f1-1-mu2-3.json)
* [network-config-f1-0.5-mu2-3.json](pruning_illustration_network/network-config-f1-0.5-mu2-3.json)
* [network-config-f1-0.25-mu2-3.json](pruning_illustration_network/network-config-f1-0.25-mu2-3.json)

### Connectivity block in config file:
```
    "Connectivity": {
        "ballandsticks,ballandsticks": {
            "GABA": {
                "conductance": [1.1e-9, 1.5e-9],
                "pruning": {
                    "f1": 0.25,
                    "softMax": 5,
                    "mu2": 3,
                    "a3": 1,
                    "distPruning": null
                }
            }
        }
    },

```

````python
pil.prune_network(pruning_config="network-config-f1-1-mu2-3.json",
                  title="f1 = 1, mu2 = 3",
                  fig_name="2-Pruning-f1-1-mu2-3.pdf")
````

## Softmax
Soft max puts a soft limit on the number of synapses connecting a neuron pair together. If there are more than softMax synapses then the probability of keeping synapses between that pair is $P = 2 \cdot s_{max}$ / ((1 + \exp(-(n_{synapses} - s_{max}) / 5)) * n_{synapses})$ where $s_{max} is the soft max.

### Configuration files:
* [network-config-softmax-3.json](pruning_illustration_network/network-config-softmax-3.json)
* [network-config-softmax-2.json](pruning_illustration_network/network-config-softmax-2.json)
* [network-config-softmax-1.json](pruning_illustration_network/network-config-softmax-1.json)

### Connectivity block in configuration file:
```
    "Connectivity": {
        "ballandsticks,ballandsticks": {
            "GABA": {
                "conductance": [1.1e-9, 1.5e-9],
                "pruning": {
                    "f1": 1,
                    "softMax": 3,
                    "mu2": null,
                    "a3": 1,
                    "distPruning": null
                }
            }
        }
    },
```

````python
pil.prune_network(pruning_config="network-config-softmax-3.json",
                  title="softmax=3",
                  fig_name="3-Pruning-softmax-3.pdf")
````

````python
pil.prune_network(pruning_config="network-config-softmax-2.json",
                  title="softmax=2",
                  fig_name="3-Pruning-softmax-2.pdf")
````

## Distance dependent examples

The distance dependent pruning density is defined by a string that is evaluated element wise by [numexpr](https://github.com/pydata/numexpr). You can include ```d``` in your equations, which is the distance from the soma in SI units (meters).

### Configuration files
* [network-config-distance-dep-v3-1.json](pruning_illustration_network/network-config-distance-dep-v3-1.json)
* [network-config-distance-dep-v3-2.json](pruning_illustration_network/network-config-distance-dep-v3-2.json)
* [network-config-distance-dep-v3-3.json](pruning_illustration_network/network-config-distance-dep-v3-3.json)

### Connecitivity block in configuration file:
```
    "Connectivity": {
        "ballandsticks,ballandsticks": {
            "GABA": {
                "conductance": [1.1e-9, 1.5e-9],
                "pruning": {
                    "f1": 1,
                    "softMax": 5,
                    "mu2": null,
                    "a3": 1,
                    "distPruning": "np.exp(-(0.5*d/60e-6)**2)"
                }
            }
        }
    },
```

````python
pil.prune_network(pruning_config="network-config-distance-dep-v3-1.json",
                  title='"exp(-(0.5*d/70e-6)^2)"',
                  fig_name="5-Pruning-distance-dep-1.pdf")
````

````python
pil.prune_network(pruning_config="network-config-distance-dep-v3-2.json",
                  title='"exp(-(d-190e-6)**2/(2*40e-6**2))"',
                  fig_name="5-Pruning-distance-dep-2.pdf")
````

````python
pil.prune_network(pruning_config="network-config-distance-dep-v3-3.json",
                  title='"1/(1+exp(-(d-280e-6)/30e-6))"',
                  fig_name="5-Pruning-distance-dep-3.pdf")
````
