# Export network connectivity

This notebook shows how to export a network.

```python
import os

network_path = os.path.join("networks","PD-example-2k")

network_path_pd0 = os.path.join(network_path, "PD0")
network_path_pd2_ref = os.path.join(network_path, "PD2-ref")
network_path_pd2 = os.path.join(network_path, "PD2")
```

```python
from snudda.utils.export_connection_matrix import SnuddaExportConnectionMatrix

exp_pd0 = SnuddaExportConnectionMatrix(in_file=network_path_pd0, out_file="export/connectivity_pd0.csv")
exp_pd2_ref = SnuddaExportConnectionMatrix(in_file=network_path_pd0, out_file="export/connectivity_pd2.csv")
exp_pd2 = SnuddaExportConnectionMatrix(in_file=network_path_pd0, out_file="export/connectivity_pd2_ref.csv")
```

## Save euclidian distance matrix

```python
exp_pd0.save_distance_matrix(plot=True)
exp_pd2_ref.save_distance_matrix()
exp_pd2.save_distance_matrix()
```

```python
axon_speed=0.8
synaptic_delay = 0  # 1e-3

exp_pd0.save_axon_dend_distance_matrix(plot=True)
exp_pd2_ref.save_axon_dend_distance_matrix()
exp_pd2.save_axon_dend_distance_matrix()
```

## Plotting comparison

Important to note that for dSPN, iSPN and FS we have used axons for touch detection, for LTS and ChIN we have probability distributions

```python
exp_pd0.plot_comparison(include_neuron_types=["dSPN", "iSPN", "FS"])
exp_pd2_ref.plot_comparison(include_neuron_types=["dSPN", "iSPN", "FS"])
exp_pd2.plot_comparison(include_neuron_types=["dSPN", "iSPN", "FS"])
```

```python
exp_pd0.plot_comparison(include_neuron_types=["LTS", "ChIN"])
exp_pd2_ref.plot_comparison(include_neuron_types=["LTS", "ChIN"])
exp_pd2.plot_comparison(include_neuron_types=["LTS", "ChIN"])
```
