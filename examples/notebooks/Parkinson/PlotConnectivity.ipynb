# Plot connectivity of network

This notebook reads in the data from multiple networks, and plots the connectivity.

```python
import os
from snudda.plotting.plot_connectivity import PlotConnectivity

network_path_wt = os.path.join("networks", "PD-example-2k", "PD0")
network_path_pd = os.path.join("networks", "PD-example-2k", "PD2")

pc = PlotConnectivity(network_path=network_path_wt, network_name="WT")
pc.add_network(network_path=network_path_pd, network_name="PD")
```

```python
axis, _ = pc.add_plot(pre_neuron_type="dSPN", post_neuron_type="iSPN", network_name="WT")
axis, _ = pc.add_plot(pre_neuron_type="dSPN", post_neuron_type="iSPN", network_name="PD", colour="blue", error_colour="lightblue", axis=axis)
pc.draw_plot(axis=axis, figure_path="test-connection-figure.png")
```

```python
pc.load_exp_data("connectivity_data.csv")
pc.add_loaded_exp_data_to_plot(phenotype="WT", data_source="Planert 2010", network_name="WT", pre_neuron_type="dSPN", post_neuron_type="dSPN")
pc.add_loaded_exp_data_to_plot(phenotype="WT", data_source="Taverna 2008", network_name="WT", pre_neuron_type="dSPN", post_neuron_type="dSPN", exp_colour="orange")

pc.add_loaded_exp_data_to_plot(phenotype="WT", data_source="Planert 2010", network_name="WT", pre_neuron_type="dSPN", post_neuron_type="iSPN")
pc.add_loaded_exp_data_to_plot(phenotype="WT", data_source="Taverna 2008", network_name="WT", pre_neuron_type="dSPN", post_neuron_type="iSPN", exp_colour="orange")

pc.add_loaded_exp_data_to_plot(phenotype="WT", data_source="Planert 2010", network_name="WT", pre_neuron_type="iSPN", post_neuron_type="dSPN")
pc.add_loaded_exp_data_to_plot(phenotype="WT", data_source="Taverna 2008", network_name="WT", pre_neuron_type="iSPN", post_neuron_type="dSPN", exp_colour="orange")

pc.add_loaded_exp_data_to_plot(phenotype="WT", data_source="Planert 2010", network_name="WT", pre_neuron_type="iSPN", post_neuron_type="iSPN")
pc.add_loaded_exp_data_to_plot(phenotype="WT", data_source="Taverna 2008", network_name="WT", pre_neuron_type="iSPN", post_neuron_type="iSPN", exp_colour="orange")

pc.draw_exp_model_data(figure_path="exp_model_plot.png")
```
