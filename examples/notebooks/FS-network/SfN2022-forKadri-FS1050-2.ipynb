```python
import os
import numpy as np

network_path = "forKadri-FS1050-1"
```

```python
from snudda.plotting import PlotInput
input_file = os.path.join(network_path, "input-spikes-oscillating-5-5-Hz.hdf5")
spi = PlotInput(input_file)
```

```python
spi.plot_input_population_unit(population_unit_id=0, num_neurons=5, neuron_type="FS", fig_size=(15,5))
```

```python
from snudda.plotting import SnuddaPlotSpikeRaster2
sim_file = os.path.join(network_path, "simulation", "output-GJ-oscillating-5-5-Hz.hdf5")
fig_path = os.path.join(network_path, "figures")
fig_file_1 = "spike-raster-osc-GJ-5-5-Hz.png"
fig_file_2 = "spike-histogram-osc-GJ-5-5-Hz.png"
spr = SnuddaPlotSpikeRaster2(network_path=network_path, simulation_file=sim_file, figure_path=fig_path)
```

```python
spr.plot_spike_raster(fig_file=fig_file_1)
spr.plot_spike_histogram(fig_file=fig_file_2)
```

```python
from snudda.plotting import SnuddaPlotSpikeRaster2
sim_file = os.path.join(network_path, "simulation", "output-noGJ-oscillating-5-5-Hz.hdf5")
fig_path = os.path.join(network_path, "figures")
fig_file_1 = "spike-raster-osc-noGJ-5-5-Hz.png"
fig_file_2 = "spike-histogram-osc-noGJ-5-5-Hz.png"
spr2 = SnuddaPlotSpikeRaster2(network_path=network_path, simulation_file=sim_file, figure_path=fig_path)
```

```python
spr2.plot_spike_raster(fig_file=fig_file_1)
spr2.plot_spike_histogram(fig_file=fig_file_2)
```

```python
count1, bin1 = spr.plot_period_histogram_mod(neuron_id=None, period=0.2, label="GJ", color="red")
count2, bin2 = spr2.plot_period_histogram_mod(neuron_id=None, period=0.2, label="NO GJ", color="black")

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot()
ax.stairs(count1, bin1, color="red", label="GJ")
ax.stairs(count2, bin2, color="black", label="no GJ")
ax.legend()
plt.show()
```
